```python
from pathlib import Path
import pandas as pd
```

```python
#importing red wine data 
data = Path('C:/Users/jmced/Desktop/project-4/Clean_Data/red_wines.csv')

df = pd.read_csv(data)
df.head()
```

```python
# changing quality to either good or bad quality. 

def rating(quality):
    if quality > 4:
        return  1  
    else:
        return 0
```

```python
#adding good_or_bad column to data frame
df['Good_or_Bad'] = df['quality'].apply(rating)
df.head()
```

```python
redwine_df = df.drop(columns="type")
redwine_df.head()
```

```python
redwine_df_1 = redwine_df.drop(columns="quality")
redwine_df_1.head()
```

```python
redwine_df_2 = redwine_df_1.dropna()
redwine_df_2.head()
```

```python
y = redwine_df_2["Good_or_Bad"]
X = redwine_df_2.drop(columns="Good_or_Bad")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=3000,
                                random_state=1)
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
results = pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
results.head(10)
```

```python
from sklearn.metrics import accuracy_score
# Display the accuracy score for the test dataset.
accuracy_score(y_test, predictions)
```
