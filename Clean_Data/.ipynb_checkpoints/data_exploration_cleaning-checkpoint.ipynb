```python
import pandas as pd
from pathlib import Path
```

```python
file_path = Path("../Resources/winequalityN.csv")
wine_db = pd.read_csv(file_path)
wine_db.head()
```

```python
wine_db.dtypes
```

```python
wine_types = wine_db.groupby('type')['quality'].count()
wine_types
```

```python
avg_by_type = wine_db.groupby('type').mean()
avg_by_type
```

```python
red_wines = wine_db[wine_db['type'] == "red"].reset_index(drop=True)
red_wines.head()
```

```python
white_wines = wine_db[wine_db['type'] == "white"].reset_index(drop=True)
white_wines.head()
```

```python
red_wines.to_csv("red_wines.csv", index=False, header=True)
white_wines.to_csv("white_wines.csv", index=False, header=True)
```
